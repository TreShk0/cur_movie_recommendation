```python
from MovieRecomendation import MovieRecommendation as mr
import pandas as pd
import numpy as np
```

```python
# Генерация случайных данных для 25 пользователей и 10 фильмов
np.random.seed(42)  # Для воспроизводимости
m = 128
n = 74
data = np.random.randint(0, 2, size=(m, n))

# Названия фильмов
movies = [f"Movie {i+1}" for i in range(n)]

# Индексы пользователей
users = [f"User {i+1}" for i in range(m)]

# Создаем DataFrame
UI_matrix = pd.DataFrame(data, index=users, columns=movies)

UI_matrix.head()  # Показать первые 5 строк
```

```python
recommender = mr(n_recommendations=5, test=False)

# Обучаем модель
recommender.fit(UI_matrix)

# Рекомендуем фильмы пользователю с индексом 0
recomended = recommender.recommend(user=0)

print(recomended, '\n', len(recomended), '\n', len(recommender.unseen))
```

```python
recommender = mr(n_recommendations=5, test=True)

# Обучаем модель
recommender.fit(UI_matrix)

# Рекомендуем фильмы пользователю с индексом 0
recomended = recommender.recommend(user=0)

print(recomended, '\n', len(recomended), '\n', len(recommender.unseen))
```
