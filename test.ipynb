```python
import numpy as np
import pandas as pd
```

```python
user_index = ['user ' + str(i) for i in range(1,21)]
movies = ['movie ' + str(i) for i in range(1,11)]

example = pd.DataFrame(np.random.randint(0,2,[20,10]),index=user_index, columns=movies)
trans_example = example.T
```

```python
trans_example['sum']=trans_example.sum(axis=1)
```

```python
example = trans_example.sort_values(by='sum', ascending=False).T
```

```python
example
```

```python
columns = example.T[example.T['sum'] > 11].T
columns = columns[:20]
columns
```

```python
R = example[:5]
R
```

```python
U = example.T[example.T['sum'] > 11].T[:5]
U
```

```python
G = np.linalg.pinv(U.to_numpy())
```

```python
print(columns.shape)
print(G.shape)
print(R.shape)
```

```python
CGR = pd.DataFrame(np.abs(np.round(columns.to_numpy()@G@R.to_numpy(), 0)), index=user_index, columns=example.columns)
```
